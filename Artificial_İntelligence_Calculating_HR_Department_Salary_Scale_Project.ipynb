In this project, the basis of artificial intelligence was created to calculate the Hr salary scale with Polynom linear regression.

Polynomial Linear Regression General Formula:

y = a + b1x + b2x^2 + b3x^3 + b4x^4 + ....... + bN*x^N

```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

#We obtain our dataset using pandas and transfer it into our dataframe object, named df.
df = pd.read_csv("polynomial.csv", sep = ";")

#To create a polynomial regression object, we invoke the PolynomialFeatures function.
#"While invoking this function, we specify the degree (N) of the polynomial.
polynomial_regression = PolynomialFeatures(degree = 2 ) #degree = N bknz. Calling the polynomial function mentioned above:
```

```python
#As seen, the data is not distributed linearly.
#If we apply linear regression to this dataset, we would observe an inappropriate prediction line.

reg = LinearRegression()
reg.fit(df[['deneyim']],df['maas'])

plt.xlabel("Years of experience")
plt.ylabel("Salary")

plt.scatter(df["deneyim"] ,df['maas'])

xaxis = df['deneyim']
yaxis = reg.predict(df[["deneyim"]])
plt.plot(xaxis , yaxis , color="green" , label="L1inear Regression")
plt.legend()
plt.show()
```

It is not appropriate to apply linear regression to this dataset; remember, for the chosen model, you should have a thorough understanding of and familiarity with the dataset.

We adapt our x-values to fit the polynomial mentioned above

1, x, x^2 (N=2) 

```python
# Calling the PolynomialFeatures function to create a polynomial regression object
# Specify the degree (N) of the polynomial while calling this function
polynomial_regression = PolynomialFeatures(degree = 4)

x_polynomial = polynomial_regression.fit_transform(df[['deneyim']])
```

```python
# Creating our regression model object, 'reg', and fitting it to x_polynomial and y-axis by calling the fit method
# Essentially, training our regression model with the existing real data
reg = LinearRegression()
reg.fit(x_polynomial,df["maas"])
```

Now that our model is ready and trained, let's see how it generates a graph based on the available data

```python
y_head = reg.predict(x_polynomial)
plt.plot(df["deneyim"] , y_head,color = 'red',label="Polynomial Regression")
#plt.plot(x_axis, y_axis, color="green", label="linear regression")
plt.legend()

#Scatter-plotting our dataset to visually inspect how well the polynomial regression fits:
plt.scatter(df['deneyim'],df['maas'])
plt.show()
```

As you can see, we can definitely say that it fits well; polynomial regression is a good choice.
